# 2. Plantea el problema de clasificación implícito en el conjunto de datos.

Basados en los datos proporcionados y su estructura, así como en el artículo de Langseth-Portinale (2007) y Sewell-Shah (1967), el problema de clasificación se puede plantear en los siguientes términos:

1. Se trata de 5 características para un total de 10,273 instancias (estudiantes).
2. Las 5 características que se pueden ver como variables aleatorias: 
    - Género (Sex), con los siguientes posibles valores:  'male' y 'female'.
    - Status socio económico (SocioEconomicStatus), con los siguientes valores posibles: 'high', 'low', 'lower_middle' y 'upper_middle'.
    - Estímulo de los padres (ParentalEncouragement), con el siguiente recorrido posible: 'high' y 'low'.
    - Coeficiente intelectual (IntelligenceQuotient), con un recorrido: 'low', 'lower_middle', 'upper_middle' y 'high'.
    - Planes para colegio (CollePlagePlans), con dos posibles valores: 'no' y 'yes'.

En este sentido, se procede a definir la red bayesiana en los términos presentados en la literatura revisada y que para tales efectos define los siguientes pasos: 

**Paso 0: Decidir qué modelar.** En este caso, es necesario definir qué incluir en el modelo. Por sus características, en este paso estamos buscando modelar la probabilidad condicional de tener planes universitarios considerando las psosibles dependencias que esta variable aleatoria tiene con respecto a las otras variables previas: Género, status socio económico, estímulo de los padre y coeficiente intelectual. 

**Paso 1: Definir las variables.** Definir cuáles son las variables aleatorias relevantes. El rango en el caso de las variables continuas y el estado en el caso de las variables discretas.

Como se adelantó previamente y basados en la literatura revisada, la variable aleatoria dependiente es discreta y se refiere a si si el entrevistado tiene o no planes universitarios, en función de las otras características descritas previamente (Género, status socio económico, coeficiente intelectual, etc.)

**Paso 2: Definir la parte cualitativa.** Definir la estructura gráfica que seguirán las dependencias de las variables (conviene pensar las ligas o bordes como causalidades). Evitar ligas o bordes que no son prácticos y estar conscientes de que una vez procesado el modelo podría ocurrir que se encuentren nuevas ligas o bordes.

Para definir la parte cualitativa, se decidió utilizar el mismo procedimiento sugerido por Langseth, H., & Portinale, L. (2007, p. 94): La arquitectura convergente, donde cada uno de las variables explicativas es un padre y la variable dependiente es el único hijo. Similar al caso de la lesión en la espalda en el artículo de Ben Gal I. (2008, p. 8).

En un análisis pormenorizado de cada variable se pueden hacer los siguiente supuestos: 

a. El Género no puede ser un padre o tener causalidad con el status económico, o con el estímulo de los padres (al menos no en esta época) o con el IQ, razón por la cual se esperaría que tenga una causalidad directa con los Planes Universitarios.

b. De igual forma, el Status Económico no tiene una relación directa con el Género (al menos no en nuestra época) o con el IQ, aunque posiblemente sí con el estímulo de los padres.

c. El IQ por su parte, podría no tener una relación directa con el género, el status socioeconómico o el estímulo de los padres.

En virtud de los supuestos anteriores, se esperaría tener una arquitectura de red bayesiana convegente como la que se muestra a continuación:

```python

from pgmpy.models import BayesianModel
from pgmpy.estimators import ParameterEstimator
from pgmpy.inference import VariableElimination
import matplotlib.pyplot as plt
import networkx as nx

# Se define la estructura del modelo bayesiano como se plantea en el punto 2 del proyecto
model = BayesianModel([('Sex', 'CollegePlans'),
                       ('SocioEconomicStatus', 'ParentalEncouragement'),
                       ('ParentalEncouragement', 'CollegePlans'),
                       ('IntelligenceQuotient', 'CollegePlans')])

# Se establece el modelo
model.fit(df)

# Se elabora la inferencia
inference = VariableElimination(model)

# Imprimimos las probabilidades de cada nodo
for node in model.nodes():
    print(f'Node: {node}')
    print(inference.query([node], evidence={}))
    print()

```

```python
# Draw the Bayesian Network with probabilities
plt.figure(figsize=(10, 6))
pos = nx.circular_layout(model)

# Draw nodes
nx.draw(model, pos, with_labels=False, node_size=5000, node_color='lightblue')

# Annotate nodes with their probabilities
for node in model.nodes():
    prob = inference.query([node], evidence={})
    label = f'{node}\n{prob.values}'
    plt.text(pos[node][0], pos[node][1], label, horizontalalignment='center', verticalalignment='center', fontsize=12)

# Draw edge labels
edge_labels = {(u, v): '' for u, v in model.edges()}
nx.draw_networkx_edge_labels(model, pos, edge_labels=edge_labels, font_color='red')
warnings.filterwarnings("ignore")

plt.title("Bayesian Network with Probabilities")
plt.axis('off')  # Hide axis
plt.show()

```

```python
# Probabilidades conjuntas 
joint_probabilities = inference.query(['Sex', 'SocioEconomicStatus', 'ParentalEncouragement', 'IntelligenceQuotient', 'CollegePlans'])

conditional_prob_sex = inference.query(['CollegePlans'], evidence={'Sex': 'female'})
conditional_prob_ses = inference.query(['CollegePlans'], evidence={'SocioEconomicStatus': 'high'})
conditional_prob_pe = inference.query(['CollegePlans'], evidence={'ParentalEncouragement': 'high'})
conditional_prob_iq = inference.query(['CollegePlans'], evidence={'IntelligenceQuotient': 'high'})

print("Probabilidades conjuntas:")
print(joint_probabilities)
print("\nProbabilidad condicional deCollegePlans dado el Sex (female):", conditional_prob_sex.values)
print("Probabilidad condicional de CollegePlans dado SocioEconomicStatus (high):", conditional_prob_ses.values)
print("Probablidad condicional de CollegePlans dado ParentalEncouragement (high):", conditional_prob_pe.values)
print("Probablidad condicional de CollegePlans dado IntelligenceQuotient (high):", conditional_prob_iq.values)
```

```python
# distribución acumulada de probabilidades
cumulative_probabilities = {}

for node in model.nodes():
    if node != 'CollegePlans':
        # Take a single value from the column for evidence
        evidence_values = {node: df[node].iloc[0]}
        cumulative_probabilities[node] = inference.query(['CollegePlans'], evidence=evidence_values).values

print("Distribución acumulada de probabilidades:")
for node, cdf in cumulative_probabilities.items():
    print(f"{node}: {cdf}")
```
